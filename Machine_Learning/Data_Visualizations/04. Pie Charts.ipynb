## Pie Charts

* Pie Charts are one of the most popular Charts used in the domain of Data Science and Business Intelligence.
* Pie Charts are beautiful and provide users to visualize data very easily.
* Pie Charts are can be used to Visualize ```Univariate``` and ```Bivariate Data```.

**In this Tutorial, we will learn,**
  * Making a simple Pie Chart.
  * Using other parameters of pie charts, to make it look more attractive.
  * Making ```Donut Charts```.
    


````python
# importing basic libraries
import numpy as np
import pandas as pd

# importing matplot for data visualization
import matplotlib.pyplot as plt
````

````python
# lets import a dataset to work with
data = pd.read_csv('employee.csv')

# lets also check the shape of the dataset
data.shape
````

````python
# lets also check the structure of the dataset

pd.set_option('max_columns', 300)
data.head()
````

````python
# lets check the documentaion of the pie chart in matplotlib
help(plt.pie)
````

### Making a Simple Pie Chart


````python
# lets check the documentataion of the value_counts function
help(pd.value_counts)
````

````python
# lets make a simple pie Chart

# we will the Business Travel Column
plt.pie(data['BusinessTravel'].value_counts(), labels = data['BusinessTravel'].value_counts().index)
plt.show()
````

### Making a Donut Chart 

````python
# we will the Business Travel Column
plt.pie(data['BusinessTravel'].value_counts(), labels = data['BusinessTravel'].value_counts().index)

# lets draw a white circle in the middle of the pie chart
my_circle=plt.Circle( (0,0), 0.7, color='white')

# adding the circle in the plot
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
````
