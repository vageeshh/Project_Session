## Exploring Scatter Plots

* Scatter Plots are one of the most popular charts in Data Science and Business Intelligence.
* Scatter Plots are used to check the relation and dependency between two different variables.
* We can Use Scatter Plots, for checking relations between multiple variables using the third variables as the size of the data points in the plot.
* Scatter PLots are also used to track down Outliers in any Dataset very easily.

**In this Tutorial, we will learn**

````python
# importing the basic libraries 
import numpy as np
import pandas as pd

# importing matplotlib for data visualization
import matplotlib.pyplot as plt
````

````python
# lets import the dataset
data = pd.read_csv('employee.csv')

# lets check the shape of the dataset
data.shape
````

````python
# lets also check the structure of the dataset
data.head()
````

````python
# lets check the documentation of scatter plot
help(plt.scatter)
````

### Bivariate Plots

* Bivariate Plots are the simple scatter plots
* Here we need two compulsory parameters i.e., data to plotted on the x axis and the data to be plotted on the y axis.

* **In the Scatter Plot below**, 
    * We are Plotting ```Age``` in the X axis.
    * We are plotting ```Monthly Rate``` in the Y axis.

````python
# lets explore a simple scatter plot

# setting the size of the plot
plt.rcParams['figure.figsize'] = (9, 5)

# setting the parameters of the plot
plt.scatter(data['Age'],
            data['MonthlyRate'])
plt.show()
````

````python
# second example for exploring scatter plot with adding labels and titles

# setting the parameters in the plota
# also, specifying a different color for the plot
plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            color = 'red')

# specifying the xlabel, ylabel, and the title of the plot
plt.ylabel('Monthly Rate')
plt.xlabel('Total Working Years')
plt.title('Comparing Monthly Wages with Experience')
plt.show()
````

### Bubble Plots (Variable of Scatter Plots)

* <p style="color: green; font-weight: bold"> This plot is Multivariate as there are three varibles involved in the plot.
* Here, We take into account the third variable as the Size of the dots or bubbles.
* In the plot below,
    * we will plot ```Total Working years``` in the X axis,
    * we will plot ```Monthly Rate``` in the Y axis,
    * we will plot ```Distance From Home``` as the size of the Data points.

````python
# second example for exploring scatter plot with adding labels and titles

# setting the size of the plot
plt.rcParams['figure.figsize'] = (16, 6)

# setting up the parameters of the plot
# specifying the third size variable in s
plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            s = data['DistanceFromHome'],
            color = 'green')

plt.ylabel('Monthly Rate')
plt.xlabel('Total Working Years')

# specifying the fontsize of the title of the plot
plt.title('Comparing Monthly Wages with Experience', fontsize = 20)
plt.show()
````

* In the above Plot, we have used Distance from home in the plot as the third variable which is determining the size of the data points in the scatter plot.
* The Bigger the size of the point, the more is the Distance from home of the employee, and the smaller the size of the point in the plot, the lesser the distance from home of the employee.
* The Insight which we can bring up from this plot is that most of the employees who are having experience 0 t0 5 years, are having longer distance from their homes to office compared to employees having more experience.
