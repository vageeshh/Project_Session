## Line Plots

* A line chart is a graphical representation of an asset's historical price action that connects a series of data points with a continuous line.

* A line chart is the most Basic type of chart that is used in almost all the Industries to Visualize the data.

* Line Charts are very popular in Financial sectors, for visualizing Time series plots, and any data over a period of time.

* Line Charts are considered to be the first choice when we have huge number of data points to show in a plot.

````python
# lets check the documentation of the line plots available in matplotlib
help(plt.plot)
````

````python
# basic line plot

plt.plot(data['Department'],
         data['Age'])
````

* ```It is very difficult to understand to understand anything from this line plot as It is puzzled.```

&nbsp;
* ```As we are trying to visualize the Ages of employees of an organization, Every Organizations Departments consists of many employees having ages from range 18 to 65, so it gets difficult to take any insight from a plot like this.```

&nbsp;
* ```For Driving some Insights, we can aggregate the ages of employees in each departments using the mean function, then we will have average age of employees in each department, which can be easily understood.```

#### Lets check the Important Parameters used in line plots

```x: The First Variable to be Plotted on the Line Chart```

```y: The Second Variable to be Plotted on the Line Chart```

``` It is not Necassary to use x and y while declaring your variables inside the plot function```

````python
# lets try to aggregate the age using mean in each dept.

x = data.pivot_table(index = 'Department',
                     values = 'Age',
                     aggfunc = 'mean').plot(kind='line')
````

* The Above Line Plot is plotted between the Various Departments of a Company and the Average Age of the Employees from each Department.
* We got the Average age of all the employees, Using the Pivot Table where we used Mean function for aggregating the ages of the employees pertaining to a Department.

## Multiple Line Plots

````python
# lets try to aggregate the age using mean in each dept.

# to know more about pivot tables, refer to Pandas->Pivot tables tutorial

x = data.pivot_table(index = 'Department',
                     values = ['Age','TotalWorkingYears'],
                     aggfunc = 'mean').plot(kind='line')
````

* In this Plot, We are using two lines in the same plot.
* The First Line, represents the average age of employees in each department
* The Second Line, represents the average Total working years of employees in each department


````python
# lets try to aggregate the age using mean in each dept.

x = data.pivot_table(index = 'Department',
                     values = ['Age','TotalWorkingYears','HourlyRate'],
                     aggfunc = 'mean').plot(kind='line')
````

* In the above plot, We have plot three different lines in the same plot
* The First plot, represents the average age of employees in each departments
* The Second plot, represents the average Total Working Years of employees in each departmens
* The Third plot, represents the average Hourly Rate of employees in each departments.

## Multiple Plots Separately

````python
# trying to plot the line plots in a different plots

data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean').plot(kind='line', color='red')

data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean').plot(kind='line', color='blue')


data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean').plot(kind='line', color='green')



data.pivot_table(index = 'Department',
                     values = ['DailyRate'],
                     aggfunc = 'mean').plot(kind='line', color='yellow')

plt.show()
````

````python
# Trying to plot line plots in different plots Using Subplots

plt.rcParams['figure.figsize'] = (12, 5)

plt.subplot(221)
x = data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean')
plt.plot(x,color='red')
plt.ylabel('Age')

plt.subplot(222)
y = data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean')
plt.plot(y, color = 'blue')
plt.ylabel('Experience')

plt.subplot(223)
z = data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean')
plt.plot(z, color ='green')
plt.ylabel('Salary')

plt.subplot(224)
a = data.pivot_table(index = 'Department',
                     values = ['PerformanceRating'],
                     aggfunc = 'mean')
plt.plot(a, color = 'yellow')
plt.ylabel('Performance')

plt.suptitle('Line Plots using Subplots')
plt.show()
````

* Visualizing the Four Different Line Plots using Subplots
* Different Plots are having different Colors
* Y label is used to represent the y axis values

````python
# lets use different patterns for the lines

plt.rcParams['figure.figsize'] = (12, 5)

plt.subplot(221)
x = data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean')
plt.plot(x, color='red')
plt.ylabel('Age')

plt.subplot(222)
y = data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean')
plt.plot(y, '-+', color = 'blue')
plt.ylabel('Experience')

plt.subplot(223)
z = data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean')
plt.plot(z, '--',  color ='green')
plt.ylabel('Salary')

plt.subplot(224)
a = data.pivot_table(index = 'Department',
                     values = ['PerformanceRating'],
                     aggfunc = 'mean')
plt.plot(a, '-*', color = 'black')
plt.ylabel('Performance')

plt.suptitle('Using Different Patterns for Different Line Plots')
plt.show()
````

* Here, We have used different patterns for all the line plots
